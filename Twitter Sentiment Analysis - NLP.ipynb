## Dataset Information

The objective of this task is to detect hate speech in tweets. For the sake of simplicity, we say a tweet contains hate speech if it has a racist or sexist sentiment associated with it. So, the task is to classify racist or sexist tweets from other tweets.

Formally, given a training sample of tweets and labels, where label '1' denotes the tweet is racist/sexist and label '0' denotes the tweet is not racist/sexist, your objective is to predict the labels on the test dataset.

For training the models, we provide a labelled dataset of 31,984 tweets. The dataset is provided in the form of a csv file with each line storing a tweet id, its label and the tweet.

## Import modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
import nltk
import warnings
%matplotlib inline

warnings.filterwarnings('ignore')
```

## Loading the dataset

```python
df = pd.read_csv('Twitter Sentiments.csv')
df.head()
```

```python
# datatype info
df.info()
```

## Preprocessing the dataset

```python
# removes pattern in the input text
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)
    return input_txt
```

```python
df.head()
```

```python
# remove twitter handles (@user)
df['clean_tweet'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
```

```python
df.head()
```

```python
# remove special characters, numbers and punctuations
df['clean_tweet'] = df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
df.head()
```

```python
# remove short words
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: " ".join([w for w in x.split() if len(w)>3]))
df.head()
```

```python
# individual words considered as tokens
tokenized_tweet = df['clean_tweet'].apply(lambda x: x.split())
tokenized_tweet.head()
```

```python
# stem the words
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda sentence: [stemmer.stem(word) for word in sentence])
tokenized_tweet.head()
```

```python
# combine words into single sentence
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = " ".join(tokenized_tweet[i])
    
df['clean_tweet'] = tokenized_tweet
df.head()
```

```python
# extract the hashtag
def hashtag_extract(tweets):
    hashtags = []
    # loop words in the tweet
    for tweet in tweets:
        ht = re.findall(r"#(\w+)", tweet)
        hashtags.append(ht)
    return hashtags    
```

```python
# extract hashtags from non-racist/sexist tweets
ht_positive = hashtag_extract(df['clean_tweet'][df['label']==0])

# extract hashtags from racist/sexist tweets
ht_negative = hashtag_extract(df['clean_tweet'][df['label']==1])
```

## Input Split

```python
# feature extraction
from sklearn.feature_extraction.text import CountVectorizer
bow_vectorizer = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')
bow = bow_vectorizer.fit_transform(df['clean_tweet'])
```

```python
# bow[0].toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(bow, df['label'], random_state=42, test_size=0.25)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
```

```python
# training
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#testing
pred = model.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
# Get predicted probabilities
pred_prob = model.predict_proba(x_test)

# Apply threshold
pred = pred_prob[:, 1] >= 0.3

# Convert boolean to integer
pred = pred.astype(int) 

# Calculate F1 score
f1 = f1_score(y_test, pred)

print("F1 Score:", f1)
```

```python
accuracy_score(y_test,pred)
```

```python
import re
from nltk.stem.porter import PorterStemmer

# Define sensitive keywords for instant detection
HATE_KEYWORDS = {
    # Violence & Threat
    'kill', 'murder', 'slaughter', 'rape', 'shoot', 'bomb', 'attack',
    'hang', 'lynch', 'stab', 'burn', 'massacre', 'beat', 'blowup',
    
    # Racial slurs / Ethnic hate
    'nigger', 'nigga', 'nigros', 'chink', 'gook', 'wetback', 'spic', 'raghead',
    'kike', 'coon', 'camel jockey', 'jungle bunny', 'monkey', 'ape',
    
    # Religious hate
    'terrorist', 'jihadist', 'infidel', 'islamist', 'zionist', 'heathen',
    'devil worshipper', 'crusader', 'satanist',
    
    # Homophobia / Transphobia
    'faggot', 'dyke', 'tranny', 'shemale', 'homo', 'queer'  # (Note: "queer" is reclaimed in some communities but often misused)
    
    # Gender hate / Misogyny
    'bitch', 'slut', 'whore', 'cunt', 'skank', 'hoe', 'witch','shut up'
    
    # Disability slurs
    'retard', 'cripple', 'dumb', 'idiot', 'moron', 'psycho', 'lunatic',
    
    # General hate/abuse
    'hate', 'filthy', 'garbage', 'vermin', 'scum', 'disease', 'plague',
    'parasite', 'cockroach', 'subhuman', 'uncivilized', 'inferior',
    
    # Extremism trigger terms
    'exterminate', 'cleanse', 'purge', 'eradicate', 'remove', 'deport',
    
    # Incitement language
    'destroy', 'obliterate', 'wipe out', 'hunt down', 'eliminate', 'crush',

    # Insults/general toxic words
    'worst', 'disgusting', 'disgrace', 'traitor', 'snake', 'backstabber'
}
```

```python
def preprocess_tweet(tweet):
    tweet = re.sub(r"@[\w]*", "", tweet)
    tweet = re.sub("[^a-zA-Z#]", " ", tweet)
    tweet = " ".join([w for w in tweet.split() if len(w) > 3])
    tweet = tweet.split()
    stemmer = PorterStemmer()
    tweet = [stemmer.stem(word.lower()) for word in tweet]
    return " ".join(tweet)

def is_hate_speech(tweet_text):
    # Check for hate keywords first
    lower_tweet = tweet_text.lower()
    if any(word in lower_tweet for word in HATE_KEYWORDS):
        return True

    # Preprocess and use model prediction
    clean_text = preprocess_tweet(tweet_text)
    vector = bow_vectorizer.transform([clean_text])
    prob = model.predict_proba(vector)[0][1]
    return prob >= 0.3

# Real-time input
tweet_input = input("Enter a tweet to classify: ")
result = is_hate_speech(tweet_input)

if result:
    print(" The tweet **might contain hate speech.**")
else:
    print(" The tweet **does not contain hate speech.**")
```
